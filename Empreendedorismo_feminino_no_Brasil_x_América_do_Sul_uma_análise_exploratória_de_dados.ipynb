# Empreendedorismo feminino no Brasil x América do Sul: uma análise exploratória de dados

Neste trabalho, vamos comparar o índice de empreendedorismo feminino no Brasil com o de outros países da América do Sul. Os países analisados são os que o dataset no permite analisar, são eles: Argentina, Bolívia, Ecuador e Peru. Além do índice de empreendedorismo feminino, olharemos para o índice geral de empreendedorismo e a participação feminina na força de trabalho.


businesswomen.jpg

O Índice de Empreendedorismo Feminino (FEI) busca identificar quais fatores permitem o florescimento de mulheres empresárias de alto potencial - mulheres que possuem e operam negócios inovadores, em expansão de mercado e voltados para a exportação. Por meio de suas atividades empresariais, as empresárias de alto potencial melhoram seu próprio bem-estar econômico e contribuem para o tecido econômico e social da sociedade. A abordagem sistemática da FEI permite a comparação entre países e benchmarking das condições diferenciadas de gênero que muitas vezes afetam o desenvolvimento do empreendedorismo feminino de alto potencial

## Sobre os dados

---

Os dados foram obtidos a partir do relatório Women Entrepreneurship Index e Global Entrepreneurship Index publicado em 2015.
A pesquisa é limitada aos países da OCDE, onde todos os dados de 2015 estão disponíveis ao mesmo tempo no banco de dados.

```
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#Creating the palette
palet = ['#999999','#999999','#008000','#999999','#999999']
```

```
df = pd.read_csv('Dataset3.csv', sep=';')
```

```
df.head()
```

```
df.Country.value_counts()
```

```
df_SA = df[df['Country'].isin(['Brazil','Argentina', 'Ecuador', 'Bolivia', 'Peru'])].copy()
df_SA.drop(columns=['No','Inflation rate'],inplace=True)
df_SA.head()
```

# Análise Exploratória de dados

---
In this exploratory analysis, we will plot the graphs with each index and then a correlation of those indexes


## ìndice de empreendedorismo feminino

**Índice de empreendedorismo feminino da GEDI** - o Índice de Empreendedorismo Feminino - mede o desenvolvimento do empreendedorismo feminino de alto potencial em todo o mundo. Definido como "inovador, em expansão de mercado e orientado para a exportação", este índice específico de gênero utiliza a estrutura, metodologia e abordagem global exclusivas da GEDI para capturar os aspectos multidimensionais do desenvolvimento empresarial. Agora em seu terceiro ano, o índice foi lançado no encontro anual de 2013 da Dell Women Entrepreneurs Network (DWEN) e coberto pela Forbes Magazine.

https://thegedi.org/research/womens-entrepreneurship-index/

```
fig, ax = plt.subplots(1,1,figsize=(8,5))
sns.barplot(data=df_SA, y='Women Entrepreneurship Index', x='Country', palette=palet)
plt.ylabel('')
plt.xticks(fontsize=13, fontfamily='serif')
plt.xlabel(' \n\n')
plt.title('Índice de empreendedorismo feminino\n', fontsize=18, fontfamily='serif', fontweight='bold')
fig.text(0.14,0.85,f'Average = {df_SA["Women Entrepreneurship Index"].mean():.1f}\n', fontweight='bold', color='darkred', fontfamily='serif')
plt.yticks(fontsize=0)


ax.annotate(f'{float(df_SA[df_SA["Country"]=="Brazil"]["Women Entrepreneurship Index"].values)}',xy=(2,(float(df_SA[df_SA['Country']=='Brazil']["Women Entrepreneurship Index"].values)+1)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Argentina"]["Women Entrepreneurship Index"].values)}',xy=(0,(float(df_SA[df_SA['Country']=='Argentina']["Women Entrepreneurship Index"].values)+1)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Bolivia"]["Women Entrepreneurship Index"].values)}',xy=(1,(float(df_SA[df_SA['Country']=='Bolivia']["Women Entrepreneurship Index"].values)+1)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Ecuador"]["Women Entrepreneurship Index"].values)}',xy=(3,(float(df_SA[df_SA['Country']=='Ecuador']["Women Entrepreneurship Index"].values)+1)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Peru"]["Women Entrepreneurship Index"].values)}',xy=(4,(float(df_SA[df_SA['Country']=='Peru']["Women Entrepreneurship Index"].values)+1)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
plt.hlines(df_SA["Women Entrepreneurship Index"].mean(),-0.5,4.5, color = 'darkred', label='Average')

plt.box(False)
plt.show()
```

Resultado: como podemos oberservar, o Brasil se encontra abaixo da média (**34.5**) com **31.1** no índice, e em quarto colocado.

## Índice de empreendedorismo

Esta mistura de atitudes, recursos e infraestrutura é conhecida como o ‘ecossistema’ do empreendedorismo. O Índice Global de Empreendedorismo é um índice anual que mede a saúde dos ecossistemas de empreendedorismo em cada um dos 137 países. Em seguida, ele classifica o desempenho entre eles. Isso fornece uma imagem de como cada país atua no contexto doméstico e internacional.
https://thegedi.org/global-entrepreneurship-and-development-index/

```
fig, ax = plt.subplots(1,1,figsize=(8,5))
sns.barplot(data=df_SA, y='Entrepreneurship Index', x='Country', palette=palet)
plt.ylabel('')
plt.xticks(fontsize=13, fontfamily='serif')
plt.xlabel(' \n\n')
plt.title('Índice de empreendedorismo\n\n', fontsize=18, fontfamily='serif', fontweight='bold')
fig.text(0.14,0.85,f'Média = {df_SA["Entrepreneurship Index"].mean():.1f}\n\n', fontweight='bold',color='darkred', fontfamily='serif')

plt.yticks(fontsize=0)

ax.annotate(f'{float(df_SA[df_SA["Country"]=="Brazil"]["Entrepreneurship Index"].values)}',xy=(2,(float(df_SA[df_SA['Country']=='Brazil']["Entrepreneurship Index"].values)+1)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Argentina"]["Entrepreneurship Index"].values)}',xy=(0,(float(df_SA[df_SA['Country']=='Argentina']["Entrepreneurship Index"].values)+1)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Bolivia"]["Entrepreneurship Index"].values)}',xy=(1,(float(df_SA[df_SA['Country']=='Bolivia']["Entrepreneurship Index"].values)+1)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Ecuador"]["Entrepreneurship Index"].values)}',xy=(3,(float(df_SA[df_SA['Country']=='Ecuador']["Entrepreneurship Index"].values)+1)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Peru"]["Entrepreneurship Index"].values)}',xy=(4,(float(df_SA[df_SA['Country']=='Peru']["Entrepreneurship Index"].values)+1)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
plt.hlines(df_SA["Entrepreneurship Index"].mean(),-0.5,4.5, color = 'darkred', label='Average')


plt.box(False)
plt.show()
```

Resultado: O Brasil, novamente, se encontra abaixo da média (**30.0**), com **25.8** pontos no índice, em último lugar.

## Taxa da participação feminina na força de trabalho

Em quase todos os países do mundo, os homens têm maior probabilidade de participar do mercado de trabalho do que as mulheres. No entanto, essas diferenças de gênero diminuíram substancialmente e, na maioria dos países ao redor do mundo, a proporção de mulheres na força de trabalho é maior hoje do que há meio século.

Para entender os determinantes dessa taxa, acesse https://ourworldindata.org/women-in-the-labor-force-determinants

```
fig, ax = plt.subplots(1,1,figsize=(8,5))
sns.barplot(data=df_SA, y='Female Labor Force Participation Rate', x='Country', palette=palet)
plt.ylabel('')
plt.xticks(fontsize=13, fontfamily='serif')
plt.yticks(fontsize=0)
plt.xlabel(' \n\n')
plt.title('Taxa da participação feminina na força de trabalho\n', fontsize=18, fontfamily='serif', fontweight='bold')
fig.text(0.14,0.85,f'Média = {df_SA["Female Labor Force Participation Rate"].mean():.1f}\n', fontweight='bold', color='darkred', fontfamily='serif')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Brazil"]["Female Labor Force Participation Rate"].values)}',xy=(2,(float(df_SA[df_SA['Country']=='Brazil']["Female Labor Force Participation Rate"].values)+2)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Argentina"]["Female Labor Force Participation Rate"].values)}',xy=(0,(float(df_SA[df_SA['Country']=='Argentina']["Female Labor Force Participation Rate"].values)+2)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Bolivia"]["Female Labor Force Participation Rate"].values)}',xy=(1,(float(df_SA[df_SA['Country']=='Bolivia']["Female Labor Force Participation Rate"].values)+2)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Ecuador"]["Female Labor Force Participation Rate"].values)}',xy=(3,(float(df_SA[df_SA['Country']=='Ecuador']["Female Labor Force Participation Rate"].values)+2)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
ax.annotate(f'{float(df_SA[df_SA["Country"]=="Peru"]["Female Labor Force Participation Rate"].values)}',xy=(4,(float(df_SA[df_SA['Country']=='Peru']["Female Labor Force Participation Rate"].values)+2)), ha='center', fontsize=12, fontfamily='serif', fontweight = 'bold')
plt.hlines(df_SA["Female Labor Force Participation Rate"].mean(),-0.5,4.5, color = 'darkred', label='Average')
plt.box(False)
plt.show()
```

Resultado: Mais um vez, vemos o Brasil abaixo da média (**59.9**), em quarto lugar com **55.9** pontos no índice

```
plt.figure(figsize=(8,6))
sns.heatmap(df_SA.corr(),annot=True)
plt.title('Correlação\n', fontsize=20, fontfamily='serif');
```

Observamos que as correlações são baixas, ou seja, os índices não acompanham um ao outro.
